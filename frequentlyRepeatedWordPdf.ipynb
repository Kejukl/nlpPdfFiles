# Repeated Nouns from PDF files

The algorithm for analysing the pdf files depends upon the objective of required data.  In the given case we are extracting frequently repeated noun from a pdf files.  The Algorith is as follow
1. Import required library
2. Read the target pdf file & Storing the text content as string called fulltext
3. NLP usage to remove stop word and to pick up nouns
4. Create a dictionary of frequently repeated words

```python
# importing all the required modules
import regex
import PyPDF2
import re
```

Reading the pdf files and storing it as an object

```python
# creating a file object 
file = open('./CS_Strengths_2018.pdf', 'rb')
# creating a pdf reader object
fileReader = PyPDF2.PdfFileReader(file)
# print the number of pages in pdf file
print('total number of pages',fileReader.numPages)
fulltext=''
for i in range(0,fileReader.numPages):
    print('processing page number',i)
    pageObj = fileReader.getPage(i)
    #use replace to remove unwanted character "\n"
    temp=pageObj.extractText().replace("\n","").replace("...","")
    #removing number patterns if needed
    pattern = '[0-9]'
    fulltext=re.sub(pattern,' ',fulltext)
    fulltext+=temp
```

NLP usage to remove stop word and to pick up nouns

```python
import spacy
import en_core_web_sm
nlp = spacy.load('en_core_web_sm')
text = nlp(u'Each of these ways gives slightly different results. For example, if you split this sentence...')
#format of text ....
#text=nlp(fulltext)
collection=[]
x=3
stopWords=['question','time','converse','company']
def cleansing(text):
    text=nlp(text)
    collection=[]
    for i in range(len(text)):
        if text[i].lower not in stopWords:
            token = text[i].pos_
            if(text[i].pos_ =='NOUN' and len(text[i])>2 ):
                print(text[i],text[i].pos_)
                collection.append(text[i].lemma_)
    return(collection)
```

```python
pos2=cleansing(fulltext)
```

Create a dictionary of frequently repeated words

```python
import operator
def collect_word_map(word_list):
    word_map = {}
    for a_word in word_list:
        word_map[a_word] = word_map.get(a_word, 0) + 1
    return word_map
    
# Let's sort the word frequency map by word counts, 
# starting from the biggest count (reverse order), 
# and print up to 10 most frequent words
word_map = collect_word_map(pos2)
sorted_map = (sorted(word_map.items(), key=operator.itemgetter(1)))[::-1]
#Printing top 15
print(sorted_map[:15])
```
